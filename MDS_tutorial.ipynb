# Tutorial: Visualizing European City Road Travel Times using Multidimensional Scaling (MDS)
(Fayyaz Minhs)

## Objective:
In this tutorial, we will explore how to use Multidimensional Scaling (MDS) to visualize pairwise road travel times between major European cities. By applying MDS, we will project the cities onto a 2D plane while preserving their pairwise distances as much as possible. The result is a clear, intuitive map that reflects their relative positions based on travel times.

## Key Concepts Covered:
* Introduction to MDS and its purpose
* Setting up a distance matrix of road travel times between cities
* Applying MDS using Pythonâ€™s scikit-learn library
* Visualizing the MDS output using matplotlib with city labels
* Understanding the non-uniqueness of MDS solutions (by running the code again)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

# Define the travel times matrix (in hours)
cities = ["Dublin", "London", "Paris", "Zurich", "Berlin", "Helsinki",
          "Moscow", "Lisbon", "Madrid", "Rome", "Istanbul", "Athens"]

travel_times = np.array([
    [0, 7.5, 14, 18, 20, 30, 40, 28, 24, 30, 40, 45],
    [7.5, 0, 6, 10, 11, 26, 34, 22, 17, 18, 30, 35],
    [14, 6, 0, 7, 11, 28, 32, 17, 13, 14, 27, 32],
    [18, 10, 7, 0, 8, 30, 30, 22, 18, 9, 24, 28],
    [20, 11, 11, 8, 0, 26, 28, 24, 20, 15, 24, 30],
    [30, 26, 28, 30, 26, 0, 12, 40, 36, 35, 40, 45],
    [40, 34, 32, 30, 28, 12, 0, 50, 46, 45, 30, 35],
    [28, 22, 17, 22, 24, 40, 50, 0, 6, 20, 35, 38],
    [24, 17, 13, 18, 20, 36, 46, 6, 0, 14, 30, 33],
    [30, 18, 14, 9, 15, 35, 45, 20, 14, 0, 20, 25],
    [40, 30, 27, 24, 24, 40, 30, 35, 30, 20, 0, 10],
    [45, 35, 32, 28, 30, 45, 35, 38, 33, 25, 10, 0]
])

# Apply Multidimensional Scaling (MDS)
mds = MDS(n_components=2, dissimilarity="precomputed")
coordinates = mds.fit_transform(travel_times)

# Plot the results
plt.figure(figsize=(10, 8))
plt.scatter(coordinates[:, 0], coordinates[:, 1], color='blue', s=100)

# Annotate each point with the city name, closer to the points
for i, city in enumerate(cities):
    plt.text(coordinates[i, 0], coordinates[i, 1], city,
             fontsize=10, ha='center', bbox=dict(facecolor='white', alpha=0.7, edgecolor='gray', boxstyle='round,pad=0.3'))

plt.title("Map of European Cities using MDS")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.grid(True)
plt.show()
```
